```python
import bagpy
import cv2
from bagpy import bagreader
import rosbag
import numpy as np
import pandas as pd
#file path of bag
bagPath = '2022-04-09-18-05-39.bag'
bag = bagreader(bagPath)
```

```python
bag.topic_table
```

```python
#location of the drone in what units?
parameterMsg = bag.message_by_topic('/drone/camera/color/camera_info')
parameterMsg
```

```python
rosbagBag = rosbag.Bag(bagPath)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, 6.0, (640,480))
for topic, msg, t in rosbagBag.read_messages(topics=['/drone/camera/color/image_raw']):
    img = np.frombuffer(msg.data, np.uint8)
    img = img.reshape(480, 640, 3)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    out.write(img)

out.release()
```
